```python
import tensorflow as tf
tf.set_random_seed(777)

x_data=[[2],[4],[6],[8],[10]]
# 0 :  , 1 : 

y_data=[[0],[0],[0],[1],[1]]
print(x_data)
print(y_data)
```

```python
X = tf.placeholder(tf.float32,shape=[None,1])
Y = tf.placeholder(tf.float32,shape=[None,1])
W = tf.Variable(tf.random_normal([1]),name='weight')
b = tf.Variable(tf.random_normal([1]),name='bias')

# 0 1 
model = tf.sigmoid(W*X+b) #    tensorflow  

# y 0 1 binary        
loss = -tf.reduce_mean(Y*tf.log(model)+(1-Y)*tf.log(1-model))
train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(loss)
print(train)
# 0~100
# 0.6

predicted = tf.cast(model>0.6, dtype=tf.float32)
# 
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y),dtype=tf.float32))

sess= tf.Session()
sess.run(tf.global_variables_initializer())
    
for step in range(10001):
    loss_val, _ = sess.run([loss,train],feed_dict={X:x_data,Y:y_data})
    if step % 200 == 0:
        print(step,loss_val)
    
```

```python
h,c,a = sess.run([model,predicted,accuracy],feed_dict={X:x_data,Y:y_data})
```

```python
print("-------------------------------")
print("model : ", h,", : ",c, ", : ",a)
print("-------------------------------")
```
