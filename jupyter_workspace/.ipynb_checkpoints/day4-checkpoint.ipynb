```python
x = np.arange(0, 3*np.pi ,0.1)
print(x)
```

```python
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)

print(y_sin)
```

```python
plt.plot(x,y_sin)
plt.plot(x,y_cos)
plt.xlabel("x축")
plt.ylabel("y축")
plt.title("sin과 cos")
plt.legend(["Sine","Cosine"])
plt.show()

plt.plot(x,y_tan)
plt.show()
```

```python
# 2행 1열로 화면에 출력
plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title("Sine")

plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title("Cosine")
plt.show()
```

```python
index = [0.7,0.6,0.3,0.9,0,0.1]
d1 = [0.4,0.51,0.37,0.56,0.68,0.02]
p1 = plt.plot(index,d1,'ro-')
d2 = [0.1,0.22,0.45,0.12,0.33,0.4]
p2 = plt.plot(index,d2,'g^-')
d3= [0.3,0.1,0.62,0.4,0.77,0.1]
p3 = plt.plot(index,d3,'bs-')
fig = plt.gcf() # 차트를 이미지로 저장 선언 
plt.legend(['p1','p2','p3'],loc=3)
plt.grid(True)
plt.show()
fig.savefig('test.png') # 저장 작업
```

```python
import numpy as np
import matplotlib.pyplot as plt
from pandas import Series
from numpy.random import randn
```

```python
# 이렇게 해주면 윈도우를 여러개 띄울수 있다.
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
fig.add_subplot(2,2,3)

# print(randn(50).cumsum())
plt.plot(randn(50).cumsum(), 'ko--', label='good')
ax1.hist(randn(100), bins=20, color='k', alpha=0.3)
ax2.scatter(np.arange(30),np.arange(30)+3*randn(30))
plt.legend(loc="best")
plt.show()
```
