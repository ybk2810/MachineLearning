numpy 배열은 동일한 자료형을 가지는 값들이 격자판 형태로 있는것
파이썬이 계산 과학 분야에 이용될때 핵심 역할을 하는 라이브러리 모음

Numpy는 배열이 생성될때 자료형을 스스로 구축한다.

    <li>배열 데이터를 디스크에 쓰거나 읽을 수 있는 도구와 메모리에 올려진 파일을 사용하는 도구</li>

```python
grades = [1, 3, -2, 4]
grades

#show_grades 함수를 정의하고 각 원소의값을 출력하는 함수를 정의

def show_grades(grades) :
    for g in grades:
        print(g,end=' ')
        
        
show_grades(grades)

def grades_sum(grades) :
    sum=0
    for i in grades:
        sum=sum+i
    return sum

grades_sum(grades)

def grades_avg(grades) :
    sum=grades_sum(grades)
    avg = sum/len(grades)   
    return avg

grades_avg(grades)

# 분산은 숫자별로 평균을 뺀 후 제곱하고 전체의 갯수로 나눗값
def grades_variance(grades) :
    ave = grades_avg(grades)
    vari = 0
    for su in grades:
        vari += (su - ave)**2
        return vari/len(grades)
    
grades_variance(grades)
```

```python
#numpy
print("numpy를 사용해서 대표값 구하기")
print("-------------------------------")
import numpy as np
print("합계:",np.sum(grades))
print("평균:",np.mean(grades))
print("분산:",np.var(grades))
print("표준편차:",np.std(grades))
```
