```python
import tensorflow as tf
tf.set_random_seed(777)
```

```python
x1_data=[73.,93.,89.,96.,73.]
x2_data=[80.,88.,91.,98.,66.]
x3_data=[75.,93.,90.,100.,70.]
y_data=[152.,185.,180.,196.,142.]

print(x1_data)
print(x2_data)
print(x3_data)
print(y_data)
```

```python
#  
x1 = tf.placeholder(tf.float32)
x2 = tf.placeholder(tf.float32)
x3 = tf.placeholder(tf.float32)
y = tf.placeholder(tf.float32)
# y = w1*x1 + w2*x2 + w3*x3 + b
#  (3 )

w1 = tf.Variable(tf.random_normal([1]),name="weight1")
w2 = tf.Variable(tf.random_normal([1]),name="weight2")
w3 = tf.Variable(tf.random_normal([1]),name="weight3")
b = tf.Variable(tf.random_normal([1]),name="bias")

model = w1*x1 + w2*x2 + w3*x3 + b
print(model)
```

```python
# cost / loss function 
loss = tf.reduce_mean(tf.square(model-y))

optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.00001)
train = optimizer.minimize(loss)
print(train)
```

```python
# session 
sess = tf.Session()
# 
sess.run(tf.global_variables_initializer())
for step in range(60001):
    loss_val, model_val, _ = sess.run(
        [loss,model,train],feed_dict={x1:x1_data,x2:x2_data,x3:x3_data,y:y_data})
    if step % 100 == 0:
        print("loss :", loss_val, " : ",model_val)
print(sess.run([w1,w2,w3,b]))        
```

```python
import tensorflow as tf
tf.set_random_seed(777)
#  : 5 3
x_data = [[73.,80.,75.],
         [93.,88.,93.],
         [89.,91.,90.],
         [96.,98.,100.],
         [73.,66.,70.]]
# 5 1
y_data = [[152.],[185.],[180.],[196.],[142.]]

print(x_data)
print(y_data)
```

```python
#  
X = tf.placeholder(tf.float32,shape=(None,3))
Y = tf.placeholder(tf.float32,shape=(None,1))
W = tf.Variable(tf.random_normal([3,1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

# matrix multiplication
# 3*5     5*?
model = tf.matmul(X,W)+b

loss = tf.reduce_mean(tf.square(model-Y))
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.00001)
train = optimizer.minimize(loss)

sess = tf.Session()
# 
sess.run(tf.global_variables_initializer())

for step in range(2000):
    loss_val , model_val, _ = sess.run([loss,model,train],feed_dict={X:x_data,Y:y_data})
    if step % 10 == 0:
        print("loss:",loss_val, ":",model_val)
        
print(sess.run([W,b]))
```

```python
# product_sales.csv  
# x_data : visit_count, buy_count
# y_data : tot_price   

# Multiple linear Model

import tensorflow as tf
tf.set_random_seed(777)
import pandas as pd
ps = pd.read_csv("product_sales.csv",sep=",")
x_data = ps[['visit_count','buy_count']]
y_data = ps[['tot_price']]
print(type(x_data))
print(type(y_data))

print(y_data.head())
```

```python
x0 = tf.placeholder(tf.float32,shape=(None,2))
y0 = tf.placeholder(tf.float32,shape=(None,1))
w0 = tf.Variable(tf.random_normal([2,1]),name='wei')
b0 = tf.Variable(tf.random_normal([1]),name="bi")

model0 = tf.matmul(x0,w0)+b0

cost = tf.reduce_mean(tf.square(model0-y0))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.00001)
train0 = optimizer.minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for step in range(8001):
    cost_val , model0_val , _ = sess.run([cost,model0,train0],feed_dict={x0:x_data,y0:y_data})
    if step % 100 == 0:
        print("cost:",cost_val,":",model0_val)
print(sess.run([w0,b0]))
```
