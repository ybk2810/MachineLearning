```python
import tensorflow as tf
tf.set_random_seed(777)

x_data=[[2],[4],[6],[8],[10]]
# 0 : 불합격 , 1 : 합격

y_data=[[0],[0],[0],[1],[1]]
print(x_data)
print(y_data)
```

```python
X = tf.placeholder(tf.float32,shape=[None,1])
Y = tf.placeholder(tf.float32,shape=[None,1])
W = tf.Variable(tf.random_normal([1]),name='weight')
b = tf.Variable(tf.random_normal([1]),name='bias')

# 0과 1로 함수
model = tf.sigmoid(W*X+b) # 자세한 내용 공식은 tensorflow 홈페이지에 있음

# y는 0과 1 binary이며 경사 하강의 연속 실행을 위해 다음과 같이 설계
loss = -tf.reduce_mean(Y*tf.log(model)+(1-Y)*tf.log(1-model))
train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(loss)
print(train)
# 0~100
# 0.6

predicted = tf.cast(model>0.6, dtype=tf.float32)
# 정확도
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y),dtype=tf.float32))

sess= tf.Session()
sess.run(tf.global_variables_initializer())
    
for step in range(10001):
    loss_val, _ = sess.run([loss,train],feed_dict={X:x_data,Y:y_data})
    if step % 200 == 0:
        print(step,loss_val)
    
```

```python
h,c,a = sess.run([model,predicted,accuracy],feed_dict={X:x_data,Y:y_data})
```

```python
print("-------------------------------")
print("model : ", h,", 예측값: ",c, ", 정확도: ",a)
print("-------------------------------")
```
