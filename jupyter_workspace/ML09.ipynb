```python
# XOR 문제 ==> 텐서 플로우로 풀어보자
# XOR = 서로 다르면 false 같으면 true
import tensorflow as tf
import numpy as np
```

```python
tf.set_random_seed(777) # random 값
learning_rate = 0.1 
x_data=[[0,0],[0,1],[1,0],[1,1]]
y_data=[[0],[1],[1],[0]]
```

```python
x_data
```

```python
X = tf.placeholder(tf.float32, shape=[None,2]) # n행 2열짜리 행렬
Y = tf.placeholder(tf.float32, shape=[None,1]) # n행 1열짜리 행렬

W = tf.Variable(tf.random_normal([2,1]), name='weight') # 가중치, X와 행렬곱을 하기 위해 [2,1]
b = tf.Variable(tf.random_normal([1]), name='bias') # 절편
```

```python
model = tf.sigmoid(tf.matmul(X,W)+b)
```

```python
cost = -tf.reduce_mean(Y*tf.log(model) + (1-Y)*tf.log(1-model)) # 포도송이
train = tf.train.GradientDescentOptimizer(learning_rate = learning_rate).minimize(cost)
```

```python
# 정확도 계산
# 0.5 보다 크면 True 아니면 False
```

```python
predicted = tf.cast(model>0.5,dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted,Y),dtype=tf.float32))
```

```python
# 세션 생성
sess = tf.Session()
# 초기화
sess.run(tf.global_variables_initializer())

for step in range(10001):
    sess.run(train, feed_dict={X:x_data,Y:y_data})
    if(step % 10  == 0):
        print(step, sess.run(cost, feed_dict={X:x_data,Y:y_data}))
        
# 정확도
h, c, a  = sess.run([model,predicted, accuracy],feed_dict={X:x_data,Y:y_data})
print("model:", h , " Correct : " , c , " accuracy : ", a)
```
